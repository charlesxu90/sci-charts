## Pie chart -- regular

```python
import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

sizes = [12, 40, 25, 15,8]
labels = ['A','B','C','D',"E"]
explode = (0, 0., 0, 0,0)  

ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=False, startangle=90, colors=plot_style_utils.CAT_PALETTE[:5],
       wedgeprops={'linewidth':.8, 'edgecolor': 'k'},
       textprops={'size': 10}
       )

ax.legend(labels, fontsize=12, handlelength=1.2, handleheight=1.2,
          loc='lower center', ncol=5, frameon=False, bbox_to_anchor=(0.5, -0.1) )

plot_style_utils.save_for_pub(fig=fig, path="figures/pie_chart", include_vector=False)
```

```python
import plot_style_utils


plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

data = [360, 158]
labels=['Within selected sites', 'Out of selected sites',]

ax.pie(data, autopct='%1.1f%%', startangle=90,
       colors=plot_style_utils.CAT_PALETTE[:len(data)], labels=None)
ax.set_title(f'R1-L1 ({sum(data)})')
ax.set_ylabel('')
ax.set_xlabel('')
ax.legend(loc='upper right', frameon=False, bbox_to_anchor=(0, 1), labels=labels)
plot_style_utils.save_for_pub(fig=fig, path=f'figures/pie_is_in_hotspot')
```
