# Bivariate distribution (regular)

```python
import plot_style_utils
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plot_style_utils.set_pub_plot_context(context="talk")
fig = plt.figure(figsize=(8, 8))

penguins = sns.load_dataset("penguins")
g = sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", 
              hue="species", s=100, alpha=0.7, palette="viridis")

g.ax_joint.legend(title="", frameon=False, markerscale=1, bbox_to_anchor=(0.05, 1.25))
g.ax_joint.set_xlabel("Bill Length (mm)")
g.ax_joint.set_ylabel("Bill Depth (mm)")

plot_style_utils.prettify_ax(g.ax_joint)
plot_style_utils.prettify_ax(g.ax_marg_x)
plot_style_utils.prettify_ax(g.ax_marg_y)

plot_style_utils.save_for_pub(fig=fig, path='figures/bivariate_jointplot')
```

# Bivariate histgram

```python
import plot_style_utils
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plot_style_utils.set_pub_plot_context(context="talk")
fig = plt.figure(figsize=(8, 8))

tips = sns.load_dataset("tips")

g = sns.jointplot(x="total_bill", y="tip", data=tips, kind="reg", 
                  scatter=False, xlim=(0, 60), ylim=(0, 12), color='k')

sns.scatterplot(data=tips, x='total_bill', y='tip',
                hue='size', palette='husl',
                size='size', sizes=(10, 200), legend='full',
                ax=g.ax_joint)
g.ax_joint.legend(title="", frameon=False, markerscale=1, bbox_to_anchor=(-0.15, 1.25))
g.ax_joint.set_xlabel("Total Bill ($)")
g.ax_joint.set_ylabel("Tip ($)")

plot_style_utils.prettify_ax(g.ax_joint)
plot_style_utils.prettify_ax(g.ax_marg_x)
plot_style_utils.prettify_ax(g.ax_marg_y)

plot_style_utils.save_for_pub(fig=fig, path='figures/bivariate_jointplot_histogram')
```
