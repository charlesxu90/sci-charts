# Violin plot

```python
import pandas as pd
import seaborn as sns

violin_data = pd.read_excel("data/villion_data.xlsx")
violin_data['values'] = violin_data['values'].astype('float')
```

## Violin plot -- regular (confidence interval inside)

```python
import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

df = violin_data
df['class'] = df['class'].astype('category')
df['values'] = df['values'].astype('float')

sns.violinplot(x="class", y="values", data=violin_data, palette=plot_style_utils.CAT_PALETTE[:3],
                   ax=ax, linewidth=2)

ax.set_xlabel("Class")

plot_style_utils.save_for_pub(fig=fig, path="figures/violinplot", include_vector=False)
```

## Violin plot -- errorbar

```python
import numpy as np
import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

df = violin_data
df['class'] = df['class'].astype('category')
df['values'] = df['values'].astype('float')

sns.violinplot(x="class", y="values", data=violin_data, palette=plot_style_utils.CAT_PALETTE[:3],
                   ax=ax, linewidth=2, inner=None)

sns.pointplot(x="class", y="values", data=violin_data, estimator="mean", errorbar='sd', err_kws={'linewidth': 2},
              linestyles="",color='k',ax=ax)

ax.set_xlabel("Class")

plot_style_utils.save_for_pub(fig=fig, path="figures/violinplot_error", include_vector=False)
```

## Violin plot -- inner box

```python
import numpy as np
import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

df = violin_data
df['class'] = df['class'].astype('category')
df['values'] = df['values'].astype('float')

sns.violinplot(x="class", y="values", data=violin_data, palette=plot_style_utils.CAT_PALETTE[:3],
                   ax=ax, linewidth=2, inner=None)

sns.boxplot(x="class",y="values",data=violin_data,color="k",width=.4,showfliers = False, showcaps = False,
            boxprops={"linewidth":1.2,"facecolor":"none",'zorder': 2},
            medianprops={"linewidth":2.2,"color":"k"},
            whiskerprops={"linewidth":2,"color":"k"},ax=ax)
ax.set_xlabel("Class")

plot_style_utils.save_for_pub(fig=fig, path="figures/violinplot_box", include_vector=False)
```
