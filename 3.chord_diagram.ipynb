```python
import numpy as np
import plot_style_utils
from mpl_chord_diagram import chord_diagram

plot_style_utils.set_pub_plot_context(context="talk")

fig, ax = plot_style_utils.simple_ax((6,4))
plot_style_utils.prettify_ax(ax)

flux_data = np.array([
    [0, 5, 6, 4, 7, 4],
    [5, 0, 5, 4, 6, 5],
    [6, 5, 0, 4, 5, 5],
    [4, 4, 4, 0, 5, 5],
    [7, 6, 5, 5, 0, 4],
    [4, 5, 5, 5, 4, 0],
])
names = ["A","B","C","D","E","F"]
chord_diagram = chord_diagram(mat=flux_data, names=names, alpha=.8, use_gradient=True, ax=ax)

plot_style_utils.save_for_pub(fig=fig, path="figures/chord_diagram")
```
