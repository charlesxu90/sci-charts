```python
import matplotlib.pyplot as plt
import plot_style_utils
import pandas as pd

labels = ["Search space", "GPT generated variants", "Wet-lab libraries", "Variants assayed in wet-lab", "Variants with activities > WT"]
vals = [975000, 77900, 58400, 3970, 250]

df = pd.DataFrame({"Labels": labels, "Values": vals})

with plt.style.context("fivethirtyeight"):
    plot_style_utils.set_pub_plot_context(context="talk")
    fig, ax = plot_style_utils.simple_ax()
    plot_style_utils.prettify_ax(ax)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)

    plt.fill_betweenx(y=[1, 3.8], x1=[9, 9.5], x2=[9, 8.5], color=plot_style_utils.CAT_PALETTE[4])
    plt.fill_betweenx(y=[4, 6.9], x1=[9.5, 10], x2=[8.5, 8], color=plot_style_utils.CAT_PALETTE[3])
    plt.fill_betweenx(y=[7, 9.9], x1=[10, 12], x2=[8, 6], facecolor=plot_style_utils.CAT_PALETTE[2])
    plt.fill_betweenx(y=[10, 12.8], x1=[12, 15], x2=[6, 3], color=plot_style_utils.CAT_PALETTE[1])
    plt.fill_betweenx(y=[13, 15.9], x1=[15, 20], x2=[3, -2], color=plot_style_utils.CAT_PALETTE[0])
    plt.xticks([],[])
    plt.yticks([2,5,8,11,14], df["Labels"][::-1],)
    for y, value in zip([2, 5, 8, 11, 14], df["Values"][::-1]):
        plt.text(9, y, value, fontsize=16, fontweight="bold", color="black", ha="center")
    plt.grid(visible=False)

plot_style_utils.save_for_pub(fig=fig, path="figures/funnel_plot_tight", include_vector=False)
```
