# Error bar

```python
import matplotlib.pyplot as plt
import numpy as np
import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

x = [2, 4, 6]
y = [4, 6, 3]

xerr = [0.5, 0.8, 1.1]
yerr = [1, 2, 1.5]

ax.errorbar(x, y, xerr=xerr, yerr=yerr, fmt='o', ecolor='k', 
            elinewidth=3, capsize=6, capthick=3, ms=10, mew=2, mfc='w', mec='k')

ax.set_ylim(0, 8)
ax.set_xlabel("Class")
ax.set_ylabel("Values")
plot_style_utils.save_for_pub(fig=fig, path="figures/errorbar_only_xyerr", include_vector=False)
```

# Dot plot

```python
import matplotlib.pyplot as plt
import numpy as np
import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
data = [3, 4, 2, 5, 3, 4, 6, 7]  # Be cautious, the value shouldn't be too large

for value, count in zip(labels, data):
    ax.plot([value]*count, range(count), marker='o', ms=12, mew=2, mec='k', mfc=plot_style_utils.CAT_PALETTE[0], color='grey', linestyle='None')
ax.grid(True)
ax.set_xlabel("Class")
ax.set_ylabel("Values")
plot_style_utils.save_for_pub(fig=fig, path="figures/dotplot", include_vector=False)
```

# Cleveland-style dot plot

```python
import numpy as np
import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

values = [3, 4, 2, 5, 3, 4, 6, 7]
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
y = np.arange(len(labels))[::-1]

# add horizontal line
ax.plot(values[::-1], y, marker='o', ms=12, mec='k', linestyle='', markeredgewidth=2, color=plot_style_utils.CAT_PALETTE[0])
ax.yaxis.set_ticks(y)
ax.yaxis.set_ticklabels(labels)

ax.tick_params(axis='y', which='major', right='on', left='on', color='0.8')
ax.grid(True, color='0.9', )

ax.set_xlabel("Values")
ax.set_ylabel("Class")
plot_style_utils.save_for_pub(fig=fig, path="figures/horizontal_dotplot", include_vector=False)
```

# Grouped barchart

```python
import pandas as pd
import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

labels = ['A', 'B', 'C', 'D', 'E']
type1 = [3, 4, 2, 5, 3]
type2 = [4, 3, 4, 6, 7]
type3 = [2, 5, 6, 3, 4]

x = np.arange(len(labels))  # the label locations

width = 0.25  # the width of the bars
ax.bar(x - width, type1, width, label='Type1', color=plot_style_utils.CAT_PALETTE[0])
ax.bar(x, type2, width, label='Type2', color=plot_style_utils.CAT_PALETTE[1])
ax.bar(x + width, type3, width, label='Type3', color=plot_style_utils.CAT_PALETTE[2])

ax.grid(True, color='0.6', linestyle='--', linewidth=1)
ax.set_xlabel("Class")
ax.set_ylabel("Values")
ax.set_ylim(0, 7.5)

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(ncol=3, loc='upper center', frameon=False, handlelength=1.5, bbox_to_anchor=(0.5, 1.2))
plot_style_utils.save_for_pub(fig=fig, path="figures/barplot_grouped", include_vector=False)
```

# Percent stacked barchart

```python
import numpy as np
import plot_style_utils

plot_style_utils.set_pub_plot_context(context="talk")
fig, ax = plot_style_utils.simple_ax()
plot_style_utils.prettify_ax(ax)

labels = ['A', 'B', 'C', 'D', 'E']
type1 = np.array([3, 4, 2, 5, 3])
type2 = np.array([4, 3, 4, 6, 7])
type3 = np.array([2, 5, 6, 3, 4])

all_data = np.array([type1, type2, type3])
sums = all_data.sum(axis=0)
bar_labels = np.array([f'Type{i+1}' for i in range(len(all_data))])

bottom_y = np.zeros(len(labels))
for i, (data, b_label) in enumerate(zip(all_data, bar_labels)):
    y = data/sums
    ax.bar(labels, y, label=b_label, color=plot_style_utils.CAT_PALETTE[i], bottom=bottom_y)
    bottom_y += y

for c in ax.containers:
    ax.bar_label(c, label_type='center', labels=[f'{100*x:.1f}' for x in c.datavalues], color='w', fontsize=14)

ax.grid(True, color='0.6', linestyle='--', linewidth=1)
ax.set_xlabel("Class")
ax.set_ylabel("Values")
ax.set_ylim(0, 1.0)

ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(ncol=3, loc='upper center', frameon=False, handlelength=1.5, bbox_to_anchor=(0.5, 1.2))
plot_style_utils.save_for_pub(fig=fig, path="figures/barplot_percent_stacked", include_vector=False)
```
